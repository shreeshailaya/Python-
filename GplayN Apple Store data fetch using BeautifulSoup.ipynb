```
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

pkgname=input('Enter package name:-  ')
blink='https://play.google.com/store/apps/details?id='
url=blink+pkgname


headers={
    "User-Agent":'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:68.0) Gecko/20100101 Firefox/68.0'
}


page=requests.get(url, headers=headers)
soup=BeautifulSoup(page.content,'html.parser')

def titleFetch():
  title=soup.title.text
  print('Title:-', title)

def decFetch(): 
  for e in soup.findAll('br'):
    e.extract()


  for f in soup.findAll('b'):
    f.extract()
  
  for g in soup.findAll('a'):
    g.extract()

  info=soup.find('div',{'jsname':'sngebd'}).text
  print('Description:-',info)
  

def imgFetch():
  image=soup.find('img',{'class':'T75of sHb2Xb'})
  imgurl=image["src"]
  img=mpimg.imread(imgurl)
  imgplot = plt.imshow(img)
  plt.show()


def authFetch():
  auth=soup.find('a',{'class':'hrTbp R8zArc'})
  #auth1=auth.get_text()
  auth1=auth.text
  print('Developer:',auth1)

def ndownldsFetch():
  ndown=soup.findAll('span',{'class':'htlgb'})[5].text
  print('No Of Downloads:- ' ,ndown)

def ratingFetch():
  rat=soup.find('div',{'class':'BHMmbe'}).text
  print('Rating:- ' ,rat)
  

def noRevFetch():
  rats=soup.find('span',{'class':'EymY4b'}).text
  print('Rating:- ' ,rats)

noRevFetch()
ratingFetch()
ndownldsFetch()
titleFetch()
authFetch()
imgFetch()
decFetch()





```
